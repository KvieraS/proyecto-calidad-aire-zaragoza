```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
df = pd.read_csv(
    "../data/processed/dataset_final_aire_meteo_zaragoza.csv",
    parse_dates=["fecha", "datetime"]
)

print("Shape inicial:", df.shape)
display(df.head())
```

```python
print("Columnas:")
print(df.columns.tolist())

print("\nTipos de datos:")
print(df.dtypes)

print("\nNulos:")
display(df.isnull().sum().sort_values(ascending=False))

print("\nDuplicados exactos:", df.duplicated().sum())
```

```python
# Ordenar dataset
df = df.sort_values(["datetime", "estacion", "contaminante"]).reset_index(drop=True)

print("Shape tras ordenar:", df.shape)
display(df.head())
```

```python
# Asegurar que fecha y datetime son de tipo fecha
df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")
df["datetime"] = pd.to_datetime(df["datetime"], errors="coerce")

# Variables derivadas útiles para análisis y dashboard
df["anio"] = df["datetime"].dt.year
df["mes_num"] = df["datetime"].dt.month
df["dia_mes"] = df["datetime"].dt.day
df["hora_real"] = df["datetime"].dt.hour
df["trimestre"] = df["datetime"].dt.quarter

print(df[["fecha", "datetime"]].dtypes)
print("Nulos en fecha:", df["fecha"].isnull().sum())
print("Nulos en datetime:", df["datetime"].isnull().sum())
display(df[["fecha", "datetime"]].head())
```

```python
print("Rango temporal:")
print(df["datetime"].min(), "->", df["datetime"].max())

print("\nContaminantes:")
print(sorted(df["contaminante"].dropna().unique().tolist()))

print("\nEstaciones:")
print(sorted(df["estacion"].dropna().unique().tolist()))
```

```python
# Guardar dataset final limpio
df.to_csv("../data/processed/dataset_final_aire_meteo_zaragoza_limpio.csv", index=False)

print("Archivo guardado:")
print("../data/processed/dataset_final_aire_meteo_zaragoza_limpio.csv")
print("Shape final limpio:", df.shape)
```
